### Use Case:

Comapare Shahid Afridi, Shane watson and Yuvraj Singh's ODI carrers. Compare how many matches they have played and how many runs they have scored in there tenure respectively.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
Players = ['S.afridi','S.Watson','Yuvi']
RunsInODIS = [8064,5757,8701]
MatPayed = [398,190,304]
```

```python
xpos = np.arange(len(Players))
xpos
```

```python
plt.xticks(xpos,Players)
plt.ylabel("Runs Scored")
plt.title("Run Scored by each indvidual in ODIs")
plt.bar(xpos, RunsInODIS)
```

```python
plt.xticks(xpos,Players)
plt.ylabel("Runs Scored")
plt.title("Run Scored by each indvidual in ODIs")
plt.bar(xpos, RunsInODIS,label="Runs Scored By each Batsmen")
plt.legend()
```

```python
plt.xticks(xpos,Players)
plt.ylabel("Runs Scored")
plt.title("Run Scored by each indvidual in ODIs")
plt.bar(xpos, RunsInODIS,label="Runs Scored By each Batsmen")
plt.bar(xpos, MatPayed,label="Matches played By each Batsmen")
plt.legend()
```

```python
#Its because xpos is same for both of these graphs to plot
```

```python
plt.xticks(xpos,Players)
plt.ylabel("Runs Scored")
plt.title("Run Scored by each indvidual in ODIs")
plt.bar(xpos-0.2, RunsInODIS,label="Runs Scored By each Batsmen")
plt.bar(xpos+0.2, MatPayed,label="Matches played By each Batsmen")
plt.legend()
```

```python
plt.xticks(xpos,Players)
plt.ylabel("Runs Scored")
plt.title("Run Scored by each indvidual in ODIs")
plt.bar(xpos-0.2, RunsInODIS,width =0.4,label="Runs Scored By each Batsmen")
plt.bar(xpos+0.2, MatPayed,width =0.4,label="Matches played By each Batsmen")
plt.legend()
plt.grid()
```

```python
#For Horizontal bar charts:

plt.yticks(xpos,Players)
#plt.ylabel("Runs Scored")
plt.title("Run Scored by each indvidual in ODIs")
plt.barh(xpos+0.2, RunsInODIS,height=0.4,label="Runs Scored By each Batsmen")
plt.barh(xpos-0.2, MatPayed,height=0.4,label="Matches played By each Batsmen")
plt.legend()
```

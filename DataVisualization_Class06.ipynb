### Save Figures created on matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
Pak_Provinces_landArea = [347190,205345 ,140914 ,74521]
Province_labels = ["Balochistan","Punjab","Sindh ","KPK"]

plt.axis("equal")
plt.pie(Pak_Provinces_landArea,labels=Province_labels,autopct='%1.2f%%',radius=2,explode=[0,0.1,0.1,0])
```

```python
#Now we need to use this pie chart in my presentation lets say, then ?
```

```python
plt.axis("equal")
plt.pie(Pak_Provinces_landArea,labels=Province_labels,autopct='%1.2f%%',radius=2,explode=[0,0.1,0.1,0])
plt.savefig("class08.png")
```

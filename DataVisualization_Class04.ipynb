## Histogram in Matplotlib

A histogram is used to summarize discrete or continuous data. In other words, it provides a visual interpretation. This requires focusing on the main points, factsof numerical data by showing the number of data points that fall within a specified range of values (called “bins”).

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
pak_Covid = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(pak_Covid,bins = 3,rwidth =0.95)
```

```python
pak_Covid = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(pak_Covid,bins = [80,100,120,140,160],rwidth =0.95,color ='g')
```

## UseCase

Pakistan have reorded [113,85,90,150,149,88,93,115,135,80,77,82,129] corona deaths from different regions of Pakistan after first case in 2020

Bangladesh have reorded [67,89,98,120,133,150,84,69,89,80,120,112,100] corona deaths from different regions of Bangladesh after first case in 2020

Build a histogram to compare both of these.

P.S: This data is not authentic it is just made up for practice purposes

```python
pak_Covid = [113,85,90,150,149,88,93,115,135,80,77,82,129]
Bang_Covid = [67,89,98,120,133,150,84,69,89,80,120,112,100]
```

```python
plt.xlabel("Days After First Covid Death")
plt.ylabel("Covid Deaths in Respective Coutries")
plt.title("Comparison of Pak and Bang on corona deaths")

plt.hist([pak_Covid,Bang_Covid],bins = [80,100,120,140,160],rwidth =0.95,color =['g','r'],label = ['Pakistan','Bangladesh'])
plt.legend()
plt.grid()
```

This comparison gives a clear picture that after 80-100 days pakistan have recorded 6 deaths while bangladesh recorded 5 deaths, and so on.

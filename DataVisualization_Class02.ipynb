### Use Case

Comapare Shahid Afridi, Shane watson and Yuvraj Singh's last 7 odis scores

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

So in out previous class we have discussed about simple 2d graphs,titles and labels. In this class we will work on axes, labels, grid, and legend. Lets start it with what you would do if you have to do comparisons.


Lets compare Shahid afridi, Shanewatson and yuvrajSingh odi carrers (Batting only).

```python
#Shahid Afridi runs in his last 7 years
S_afridi_Runs = [313,601,462,182,441,354,194]

S_Watson_Runs = [1013,771,1139,377,693,165,336]

Y_Singh_Runs = [893,783,349,453,2,276,372]

Last7Years = [1,2,3,4,5,6,7]
```

```python
plt.plot(Last7Years,S_afridi_Runs)
```

```python
plt.plot(Last7Years,S_afridi_Runs)
plt.plot(Last7Years,S_Watson_Runs)
plt.plot(Last7Years,Y_Singh_Runs)
```

```python
plt.xlabel("Last 7 years of ODIS")
plt.ylabel("Runs Scored")
plt.title("Run Comparison of Cricketing legends")

plt.plot(Last7Years,S_afridi_Runs,label = "S.Afridi Score")
plt.plot(Last7Years,S_Watson_Runs,label = "S.Watson Score")
plt.plot(Last7Years,Y_Singh_Runs,label = "Yuvi Score")
```

```python
plt.xlabel("Last 7 years of ODIS")
plt.ylabel("Runs Scored")
plt.title("Run Comparison of Cricketing legends")

plt.plot(Last7Years,S_afridi_Runs,label = "S.Afridi Score")
plt.plot(Last7Years,S_Watson_Runs,label = "S.Watson Score")
plt.plot(Last7Years,Y_Singh_Runs,label = "Yuvi Score")

plt.legend()
```

```python
plt.xlabel("Last 7 years of ODIS")
plt.ylabel("Runs Scored")
plt.title("Run Comparison of Cricketing legends")

plt.plot(Last7Years,S_afridi_Runs,label = "S.Afridi Score")
plt.plot(Last7Years,S_Watson_Runs,label = "S.Watson Score")
plt.plot(Last7Years,Y_Singh_Runs,label = "Yuvi Score")

#plt.legend(loc = "best")
#plt.legend(loc = "lower left")
#plt.legend(loc = "best",shadow = True)
plt.legend(loc = "best",shadow = True,fontsize ="large") #Or small
```
